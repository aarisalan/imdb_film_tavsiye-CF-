# Kütüphanelerin İçeri Aktarılması

```python
import numpy as np
import pandas as pd
```

#  User Veri Setinin Düzenlenerek İçeri Aktarılması

```python
column_names = ["user_id", "item_id", "rating", "timestamp"]
df = pd.read_csv("users.data", sep = "\t", names = column_names)
```

```python
df.head()
```

```python
len(df)
```

# Movies Veri Setinin Düzenlenerek İçeri Aktarılması

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
len(movie_titles)
```

# Veri Setlerinin item_id Parametresi Üzerinden Birleştirilmesi

```python
#tekrar çalıştırma ikinci kere birleştirir!!!
df = pd.merge(df, movie_titles, on = "item_id")
```

```python
df
```

```python
moviemat = df.pivot_table(index = "user_id",columns ="title", values = "rating")
moviemat
```

```python
type(moviemat)
```

## Seçilen Değişkenin İncelenmesi

```python
startrek_user_ratings = moviemat["Star Trek: First Contact (1996)"]
```

```python
startrek_user_ratings
```

```python
startrek_user_ratings.mean()
```

# Seçilen Değişkenin Diğer Değişkenlerle Korelasyon Katsayısının Belirlenmesi

```python
#hata boş verilerin olmasından kaynaklanıyor
similiar_to_startrek = moviemat.corrwith(startrek_user_ratings)
```

```python
similiar_to_startrek
```

```python
type(similiar_to_startrek)
```

```python
corr_startrek = pd.DataFrame(similiar_to_startrek, columns = ["Correlation"])
corr_startrek.dropna(inplace = True)
corr_startrek.head()
```

```python
len(corr_startrek)
```

# Korelasyon Katsayısına Göre Filmlerin Sıralanması

```python
corr_startrek.sort_values("Correlation",ascending = False).head(10)
```

## Az değerlendirmeye sahip filmler yanıltıcı bir sonuç oluşturuyor. Bu sebeple bu verilerin temizlenmesi gerekiyor.

```python
df.head()
```

```python
df.drop(["timestamp"],axis = 1)
```

```python
ratings = pd.DataFrame(df.groupby("title")["rating"].mean())

ratings.sort_values("rating", ascending = False).head()
```

```python
ratings["rating_oy_sayisi"] = pd.DataFrame(df.groupby("title")["rating"].count())
ratings.head()
```

```python
ratings.sort_values("rating_oy_sayisi", ascending = False)
```

```python
corr_startrek = corr_startrek.join(ratings["rating_oy_sayisi"])
corr_startrek.head()
```

# Rating Oy Sayısı 100'den Fazla Olan Filmlerin Korelasyon Sırasına Göre Sıralanması

```python
corr_startrek[corr_startrek["rating_oy_sayisi"]>100].sort_values("Correlation", ascending = False).head(15)
```

### Bu şekilde filmler daha anlamlı halde önerilmeye başladı. Star Trek'in devam filmlerinin en üst sıralarda yer alması, bize doğru sonuç aldığımızı işaret ediyor.
